```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('../raw_data/calendar.csv')
```

```python
df.head(4)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['Month'] = df['date'].dt.month
```

```python
df['Year'] = df['date'].dt.year
```

```python
df['occupied'] = df['available'].apply(lambda x: 1 if x == 'f' else 0)
```

```python
df['next_6_month'] = df['date'].apply(lambda x: 1 if x < datetime(2020, 5, 1) and x > datetime(2019, 12, 1) else 0)
```

```python
df = df[df['next_6_month'] == 1]
```

```python
df_agg = df.groupby(by='listing_id', as_index=False).agg({'occupied': 'mean'})
```

```python
df_agg.to_csv('../generated_data/6month_occupancy.csv', index=False)
```
